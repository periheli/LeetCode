```python
from typing import List, Dict, Optional, Set, Tuple
from collections import deque, Counter
from sortedcontainers import SortedList

# support for maintaining a list in sorted order
# without having to sort the list after each insertion
import bisect
from heapq import heappop, heappush


class Solution:
    def getMaximumGold(self, grid: List[List[int]]) -> int:
        m = len(grid)
        n = len(grid[0])

        def num_gold(cell: Tuple[int, int]):
            return grid[cell[0]][cell[1]]

        def adjacent_cells(cell: Tuple[int, int]):
            row_idx = cell[0]
            col_idx = cell[1]
            cells = []
            if m > 1:
                if row_idx == 0:
                    cells.append((1, col_idx))
                elif row_idx == m - 1:
                    cells.append((m - 2, col_idx))
                else:
                    cells.extend([(row_idx - 1, col_idx), (row_idx + 1, col_idx)])
            if n > 1:
                if col_idx == 0:
                    cells.append((row_idx, 1))
                elif col_idx == n - 1:
                    cells.append((row_idx, n - 2))
                else:
                    cells.extend([(row_idx, col_idx - 1), (row_idx, col_idx + 1)])
            return cells

        def helper(start_cell: Tuple[int, int], visited_cells: Set[Tuple[int, int]]):
            print(start_cell, visited_cells)
            if num_gold(start_cell) == 0:
                return 0
            possible_next_cells = [
                cell
                for cell in adjacent_cells(start_cell)
                if num_gold(cell) > 0 and cell not in visited_cells
            ]
            print("pos_celss: ", possible_next_cells)
            max_gold = 0
            for cell in possible_next_cells:
                max_gold = max(max_gold, helper(cell, visited_cells.union({cell})))
            print("max_gold: ", max_gold)
            return num_gold(start_cell) + max_gold

        res = 0
        for row_idx in range(m):
            for col_idx in range(n):
                print("====", (row_idx, col_idx))
                res = max(res, helper((row_idx, col_idx), {(row_idx, col_idx)}))
                print(res)
        return res

    def getMaximumGold(self, grid: List[List[int]]) -> int:
        m, n = len(grid), len(grid[0])
        DIR = [0, 1, 0, -1, 0]

        def dfs(r, c):
            if r < 0 or r == m or c < 0 or c == n or grid[r][c] == 0:
                return 0
            ans = 0
            # orgGold = grid[r][c]
            # grid[r][c] = 0
            grid[r][c] = -grid[r][c]
            for i in range(4):
                ans = max(ans, dfs(r + DIR[i], c + DIR[i + 1]))
            # grid[r][c] = orgGold
            grid[r][c] = -grid[r][c]
            return ans + grid[r][c]

        ans = 0
        for r in range(m):
            for c in range(n):
                ans = max(ans, dfs(r, c))
        return ans
```
