```python
from typing import List, Dict, Optional, Set, Tuple
from collections import deque, Counter
from sortedcontainers import SortedList

# support for maintaining a list in sorted order
# without having to sort the list after each insertion
import bisect
from heapq import heappop, heappush


class Solution:
    def minMovesToSeat(self, seats: List[int], students: List[int]) -> int:
        return sum(
            abs(seat - student)
            for seat, student in zip(sorted(seats), sorted(students))
        )

    def minMovesToSeat2(self, seats: List[int], students: List[int]) -> int:
        sorted_seats, sorted_students = sorted(seats), sorted(students)
        return sum(
            abs(seat - student)
            for seat, student in zip(sorted_seats, sorted_students)
        )

    def minMovesToSeat3(self, seats: List[int], students: List[int]) -> int:
        counter = [0] * 101
        for seat in seats:
            counter[seat] += 1
        for student in students:
            counter[student] -= 1

        print(counter)
        diff = 0
        pos_sum = 0
        neg_sum = 0
        res = 0
        for i, count in enumerate(counter):
            print(i)
            print(pos_sum)
            print(neg_sum)
            prev_diff = diff
            diff += count
            print("diff: ", diff)
            sign = prev_diff * diff
            if sign > 0 or prev_diff == 0:
                if count > 0:
                    pos_sum += count * i
                else:
                    neg_sum += count * i
            else:
                if count > 0:
                    pos_sum += -prev_diff * i
                    res += abs(pos_sum + neg_sum)
                    pos_sum = diff * i
                    neg_sum = 0
                else:
                    neg_sum += -prev_diff * i
                    res += abs(pos_sum + neg_sum)
                    neg_sum = diff * i
                    pos_sum = 0
            print(pos_sum)
            print(neg_sum)
            print("res:", res)
            print("---")
        return res
```
