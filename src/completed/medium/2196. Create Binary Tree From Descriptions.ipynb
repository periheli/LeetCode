```python
from typing import List, Dict, Optional, Set, Tuple
from collections import deque, Counter
from sortedcontainers import SortedList
import random

# support for maintaining a list in sorted order
# without having to sort the list after each insertion
import bisect
from heapq import heappop, heappush, heapify, heapreplace, nlargest


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


class Solution:
    def createBinaryTree(
        self, descriptions: List[List[int]]
    ) -> Optional[TreeNode]:
        node_dict: Dict[int, TreeNode] = {}
        head_candidates: Set[int] = set()
        for parent, child, is_left in descriptions:
            if parent not in node_dict:
                node_dict[parent] = TreeNode(parent)
                head_candidates.add(parent)
            parent_node = node_dict[parent]

            if child not in node_dict:
                node_dict[child] = TreeNode(child)
            head_candidates.discard(child)
            child_node = node_dict[child]

            if is_left:
                parent_node.left = child_node
            else:
                parent_node.right = child_node

        return node_dict[head_candidates.pop()]

    def createBinaryTree_2(
        self, descriptions: List[List[int]]
    ) -> Optional[TreeNode]:
        # Best
        node_dict: Dict[int, TreeNode] = {}
        child_set: Set[int] = set()
        for parent, child, is_left in descriptions:
            parent_node = node_dict.setdefault(parent, TreeNode(parent))
            child_node = node_dict.setdefault(child, TreeNode(child))
            if is_left:
                parent_node.left = child_node
            else:
                parent_node.right = child_node

            child_set.add(child)

        return node_dict[(set(node_dict.keys()) - child_set).pop()]

    def createBinaryTree_3(
        self, descriptions: List[List[int]]
    ) -> Optional[TreeNode]:
        node_dict: Dict[int, TreeNode] = {}
        child_set: Set[int] = set()
        for parent, child, is_left in descriptions:
            if parent not in node_dict:
                node_dict[parent] = TreeNode(parent)
            parent_node = node_dict[parent]
            if child not in node_dict:
                node_dict[child] = TreeNode(child)
            child_node = node_dict[child]
            if is_left:
                parent_node.left = child_node
            else:
                parent_node.right = child_node

            child_set.add(child)

        return node_dict[(set(node_dict.keys()) - child_set).pop()]
```
