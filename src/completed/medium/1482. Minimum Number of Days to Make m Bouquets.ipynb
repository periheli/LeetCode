```python
from typing import List, Dict, Optional, Set, Tuple
from collections import deque, Counter
from sortedcontainers import SortedList
import random

# support for maintaining a list in sorted order
# without having to sort the list after each insertion
import bisect
from heapq import heappop, heappush, heapify, heapreplace, nlargest


class Solution:
    def minDays(self, bloomDay: List[int], m: int, k: int) -> int:
        n = len(bloomDay)
        if n < m * k:
            return -1

        def is_valid(day: int):
            adj_counter = 0
            num_bouquets = 0
            for i in bloomDay:
                if i <= day:
                    adj_counter += 1
                    if adj_counter == k:
                        num_bouquets += 1
                        if num_bouquets == m:
                            return True
                        adj_counter = 0
                else:
                    adj_counter = 0
            return False

        low = min(bloomDay)
        high = max(bloomDay)
        while low < high:
            mid = (low + high) // 2
            if is_valid(mid):
                high = mid
            else:
                low = mid + 1
        return low

    def minDays2(self, bloomDay: List[int], m: int, k: int) -> int:
        n = len(bloomDay)
        if n < m * k:
            return -1

        def feasible(day: int):
            adj_counter = 0
            num_bouquets = 0
            for i in bloomDay:
                if i <= day:
                    num_bouquets += (adj_counter + 1) // k
                    adj_counter = (adj_counter + 1) % k
                    if num_bouquets == m:
                        return True
                else:
                    adj_counter = 0
            return False

        low = min(bloomDay)
        high = max(bloomDay)
        while low < high:
            mid = (low + high) // 2
            if feasible(mid):
                high = mid
            else:
                low = mid + 1
        return low
```
