```python
from typing import List, Dict, Optional, Set, Tuple
from collections import deque, Counter
from sortedcontainers import SortedList

# support for maintaining a list in sorted order
# without having to sort the list after each insertion
import bisect
from heapq import heappop, heappush


class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:
        n = len(grid)
        m = len(grid[0])

        def get_connected_lands(cell: Tuple[int, int]):
            if cell[0] == 0:
                if cell[1] == 0:
                    connected_cells = []
                else:
                    connected_cells = [(cell[0], cell[1] - 1)]
            else:
                if cell[1] == 0:
                    connected_cells = [(cell[0] - 1, cell[1])]
                else:
                    connected_cells = [
                        (cell[0] - 1, cell[1]),
                        (cell[0], cell[1] - 1),
                    ]

            return [
                cell
                for cell in connected_cells
                if grid[cell[0]][cell[1]] == "1"
            ]

        connected_islands_set: Set[Tuple[int, int]] = set()
        islands = [[0] * m for _ in range(n)]
        num_islands = 0
        for row_idx in range(n):
            for col_idx in range(m):
                if grid[row_idx][col_idx] == "1":
                    connected_lands = get_connected_lands((row_idx, col_idx))
                    if len(connected_lands) == 0:
                        num_islands += 1
                        islands[row_idx][col_idx] = num_islands
                    else:
                        island = islands[connected_lands[0][0]][
                            connected_lands[0][1]
                        ]
                        if len(connected_lands) == 2:
                            connected_island = islands[connected_lands[1][0]][
                                connected_lands[1][1]
                            ]
                            if island < connected_island:
                                connected_islands_set.add(
                                    (island, connected_island)
                                )
                                islands[row_idx][col_idx] = island
                            elif island > connected_island:
                                connected_islands_set.add(
                                    (connected_island, island)
                                )
                                islands[row_idx][col_idx] = connected_island
                            else:
                                islands[row_idx][col_idx] = island
                        else:
                            islands[row_idx][col_idx] = island

        for row in islands:
            print(row)
        print(connected_islands_set)
        island_parent_dict: Dict[int, int] = {
            i: i for i in range(1, num_islands + 1)
        }

        def union_sets(elem_1: int, elem_2: int):
            def find_set(elem: int):
                if island_parent_dict[elem] == elem:
                    return elem
                return find_set(island_parent_dict[elem])

            set_1 = find_set(elem_1)
            set_2 = find_set(elem_2)
            if set_1 < set_2:
                island_parent_dict[set_2] = set_1
                return 1
            elif set_1 > set_2:
                island_parent_dict[set_1] = set_2
                return 1
            else:
                return 0

        num_sets = num_islands
        for island_1, island_2 in connected_islands_set:
            num_sets -= union_sets(island_1, island_2)

        return num_sets

    def numIslands(self, grid: List[List[str]]) -> int:
        n = len(grid)
        m = len(grid[0])

        def dfs(row_idx: int, col_idx: int):
            if (
                row_idx < 0
                or col_idx < 0
                or row_idx >= n
                or col_idx >= m
                or grid[row_idx][col_idx] == "0"
            ):
                return
            grid[row_idx][col_idx] = "0"
            dfs(row_idx - 1, col_idx)
            dfs(row_idx + 1, col_idx)
            dfs(row_idx, col_idx - 1)
            dfs(row_idx, col_idx + 1)

        res = 0
        for row_idx in range(n):
            for col_idx in range(m):
                if grid[row_idx][col_idx] == "1":
                    res += 1
                    dfs(row_idx, col_idx)

        return res
```
