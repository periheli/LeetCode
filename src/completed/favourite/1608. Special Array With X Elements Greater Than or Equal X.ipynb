```python
from typing import List, Dict, Optional, Set, Tuple
from collections import deque, Counter
from sortedcontainers import SortedList
import random

# support for maintaining a list in sorted order
# without having to sort the list after each insertion
import bisect
from heapq import (
    heappop,
    heappush,
    heapify,
    heapreplace,
    _heappop_max,
    _heapify_max,
)


class Solution:
    def specialArray(self, nums: List[int]) -> int:
        n = len(nums)
        sorted_nums = sorted(nums, reverse=True)
        if sorted_nums[-1] >= n:
            return n

        for i in range(n - 1):
            if sorted_nums[i] < i + 1:
                return -1
            if sorted_nums[i] != sorted_nums[i + 1]:
                if sorted_nums[i + 1] < i + 1:
                    return i + 1
        return -1  # test case: [0]

    def specialArray_99_99(self, nums: List[int]) -> int:
        n = len(nums)
        sorted_nums = sorted(nums, reverse=True)
        if sorted_nums[-1] >= n:
            return n

        for i in range(n - 1):
            if sorted_nums[i] < i + 1:
                return -1
            if sorted_nums[i + 1] < i + 1:
                return i + 1
        return -1

    def specialArray2(self, nums: List[int]) -> int:
        """
        Counting Sort
        """
        counter = [0] * (max(nums) + 1)
        for num in nums:
            counter[num] += 1

        acc = 0
        for i in range(len(counter) - 1, -1, -1):
            acc += counter[i]
            if acc == i:
                return i
            if acc > i:
                return -1
        # return -1

    def specialArray3(self, nums: List[int]) -> int:
        """
        Counting Sort with optimized counter length
        """
        n = len(nums)
        counter = [0] * (n + 1)
        for num in nums:
            counter[min(num, n)] += 1

        acc = 0
        for i in range(len(counter) - 1, -1, -1):
            acc += counter[i]
            if acc == i:
                return i
            if acc > i:
                return -1

    def specialArray4(self, nums: List[int]) -> int:
        """
        Sort `nums` then Binary Search
        """
        n = len(nums)
        if n == 1:
            if nums[0] >= 1:
                return 1
            else:
                return -1

        sorted_nums = sorted(nums, reverse=True)
        if sorted_nums[-1] >= n:
            return n

        left, right = 0, n - 1
        while left <= right:
            mid = (left + right) // 2
            if sorted_nums[mid] >= mid + 1:
                if sorted_nums[mid + 1] < mid + 1:
                    return mid + 1
                left = mid + 1
            else:
                right = mid - 1

        return -1
```
